```python
import os
```

```python
# Define paths
dataset_path = '/content/drive/MyDrive/dataset1'
images_train_path = os.path.join(dataset_path, 'images', 'train')
images_val_path = os.path.join(dataset_path, 'images', 'val')
labels_train_path = os.path.join(dataset_path, 'labels', 'train')
labels_val_path = os.path.join(dataset_path, 'labels', 'val')
data_yaml_path = os.path.join(dataset_path, 'data.yaml')
```

```python
# Create data.yaml (if it doesn't exist)
if not os.path.exists(data_yaml_path):
    with open(data_yaml_path, 'w') as f:
        f.write(f"train: {images_train_path}\n")
        f.write(f"val: {images_val_path}\n")
        f.write("nc: 1\n")  # Replace with your number of classes
        f.write("names: ['speed_breaker']\n")  # Replace with your class names
```

```python
# Load a model
model = YOLO('yolov8x.pt')
```

```python
# Train the model using the data.yaml file
results = model.train(data=data_yaml_path, epochs=150, imgsz=640)
```

```python
# Export the model
path = model.export(format='torchscript')  # Export to TorchScript format
```

```python
#Optional: Inference example:
results = model('/content/test.png', conf=0.35) #path to image for inference
```
